# Parte 2: Selección y limpieza de los Datos en Python

## Librerías

```python
import pandas as pd
```

## Cargar Datos

```python
df = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv')
df
```

## Preparar Datos

### Renombrar Columnas

```python
df = df.rename(columns = {'X1':'Given_Credit', 'X2':'Gender', 'X3':'Education', 'X4':'Marital_Status', 'X5':'Age', 'X6':'Repayment_Status_September_2005', 'X7':'Repayment_Status_August_2005', 'X8':'Repayment_Status_July_2005',
                          'X9':'Repayment_Status_June_2005', 'X10':'Repayment_Status_May_2005', 'X11':'Repayment_Status_April_2005', 'X12':'Bill_Statement_September_2005', 'X13':'Bill_Statement_August_2005',
                          'X14':'Bill_Statement_July_2005', 'X15':'Bill_Statement_June_2005', 'X16':'Bill_Statement_May_2005', 'X17':'Bill_Statement_April_2005', 'X18':'Paid_September_2005', 'X19':'Paid_August_2005',
                          'X20':'Paid_July_2005' ,'X21':'Paid_June_2005' ,'X22':'Paid_May_2005' ,'X23':'Paid_April_2005', 'Y':'Credible_Client'})
df
```

### Cambiar valores numéricos a categóricos

```python
df['Gender'] = df.Gender.map({1.0:'Male', 2.0:'Female'})
df['Education'] = df.Education.map({1.0:'Graduate_School', 2.0:'University', 3.0:'High_School', 4.0:'Others'})
df['Marital_Status'] = df.Marital_Status.map({1.0:'Married', 2.0:'Singel', 3.0:'Others'})
df['Credible_Client'] = df.Credible_Client.map({1.0:'Yes', 0.0:'No'})
df
```

### Eliminar columna de ID y filas con datos faltantes

```python
df.isna().sum()
```

```python
df = df.drop(['ID'], axis = 1).dropna().reset_index(drop = True)
df
```

```python
df.isna().sum()
```

### Cambiar tipo de columna

```python
df['Age'] = df.Age.astype(int)
df
```
